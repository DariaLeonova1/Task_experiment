 ### Функция:
$$\frac { \cos(x) - \sin(x) } {(1+ \sin(x) )^2} $$ Отрезок:[0, П/2]


```python
x = np.linspace(-1, 10000, 50)                   
y = (np.cos(x)-np.sin(x))/(1+np.sin(x))**2 

plt.style.use('seaborn')                        
fig, ax = plt.subplots()   
ax.set_title('График функции', fontsize = 18)
ax.plot(x, y, color="blue", label="y(x)")     
ax.set_xlabel("x", fontsize = 12)                              
ax.set_ylabel("Значения функции", fontsize = 12)                             
ax.legend()                                    

plt.show()                
```

# Минимальное, максимальное и среднее значения

```python
a=min(y)
b=max(y)
print('min_value=', a)
print('max_value=', b)
c=(a+b)/2
print('average_value=',c)
```

# Производная функции

```python
def diff(x, y):
    len_arr = len(y)
    if len(x) != len_arr:
        return None
    array_diff = np.zeros(len_arr)
    for i in range(len_arr-1):
        array_diff[i] = (y[i+1] - y[i])/(x[i+1]-x[i])
    array_diff[len_arr-1] = array_diff[len_arr - 2]
    return array_diff
        
Dif = diff(x,y)

plt.style.use('seaborn')    
fig, ax = plt.subplots()
ax.set_title('Производная функции', fontsize = 18)
ax.set_xlabel("x", fontsize = 12)
ax.set_ylabel("Производная", fontsize = 12)
ax.plot(x,Dif, color="red")

plt.show()
```
